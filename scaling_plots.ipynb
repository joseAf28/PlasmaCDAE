```python

####* Basline comparison with other scalling models and cdae dim increase


params_hidden_dim_map_augmented = params_hidden_dim_map + 23548

plt.figure()
plt.plot(params_regressor, rmse_mean_regressor, marker='o', label='Direct Regressor', color='blue')
plt.plot(params_hidden_dim_map_augmented, rmse_mean_hidden_dim_refined, marker='o', label='CDAE - Map Net', color='orange')
plt.plot(params_cdae_dim_all, rmse_cade_dim_refine, marker='o', label='CDAE - Core Net', color='green')
plt.xlabel('Number of Parameters')
plt.ylabel('RMSE')
plt.legend()
plt.savefig('figures/2-rmse_vs_params_comparison.png')


interp_regressor = sp.interpolate.interp1d(params_regressor, rmse_mean_regressor, kind='linear', fill_value="extrapolate")
interp_hidden_dim_map = sp.interpolate.interp1d(params_hidden_dim_map_augmented, rmse_mean_hidden_dim_refined, kind='linear', fill_value="extrapolate")
interp_cdae_dim = sp.interpolate.interp1d(params_cdae_dim_all, rmse_cade_dim_refine, kind='linear', fill_value="extrapolate")

x_init = max(params_regressor[0], params_hidden_dim_map_augmented[0])
x_final = min(params_regressor[-1], params_hidden_dim_map_augmented[-1])

x = np.linspace(x_init, x_final, 1000)
y_regressor = interp_regressor(x)
y_hidden_dim_map = interp_hidden_dim_map(x)

x_init2 = max(params_cdae_dim_all[0], params_regressor[0])
x_final2 = min(params_cdae_dim_all[-1], params_regressor[-1])
x2 = np.linspace(x_init2, x_final2, 1000)

y_cdae_dim2 = interp_cdae_dim(x2)
y_regressor2 = interp_regressor(x2)


plt.figure()
plt.plot(x, np.abs(y_hidden_dim_map-y_regressor)/y_regressor*100, label='Gain CDAE - Map Net', color='orange')
plt.plot(x2, (y_regressor2-y_cdae_dim2)/y_regressor2*100, label='Gain CDAE - Core Net', color='green')
plt.grid()
plt.legend()
plt.xlabel('Number of Parameters')
plt.ylabel('Gain %')
plt.savefig('figures/2-gain_vs_params_comparison.png')
```
